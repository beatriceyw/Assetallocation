> MBB 기대수익률


```python
import numpy as np
import pandas as pd
import yfinance as yf
import plotly.express as px
from pypfopt.expected_returns import mean_historical_return
```

```python
tickers = ['SPY', 'MBB']

pxclose = pd.DataFrame()

for t in tickers:
    pxclose[t] = yf.download(t, interval='1d', start = "2010-01-01")['Adj Close'].round(2)
    
# SPY 수익률 계산(10년)    
start_price_spy = pxclose['SPY'][0]
end_price_spy = pxclose['SPY'][-1]
total_return_spy = round((end_price_spy - start_price_spy) / start_price_spy, 2)

# SPY 연평균 수익률 계산
num_years_spy = len(pxclose['SPY']) / 252 # 1년은 주식 시장의 거래일인 252일로 가정
average_annual_return_spy = round((1 + total_return_spy) ** (1 / num_years_spy) - 1, 2)

# MBB 수익률 계산(10년)
start_price_mbb = pxclose['MBB'][0]
end_price_mbb = pxclose['MBB'][-1]
total_return_mbb = round((end_price_mbb - start_price_mbb) / start_price_mbb, 2)

# MBB 연평균 수익률 계산
num_years_mbb = len(pxclose['MBB']) / 252 # 1년은 주식 시장의 거래일인 252일로 가정
average_annual_return_mbb = round((1 + total_return_mbb) ** (1 / num_years_mbb) - 1, 2)

fig = px.line(pxclose)

fig.update_layout(
    title="SPY(S&P500) / MBB(모기지) expected_return",
    yaxis_title="returns($)"
)
fig.show()

# 연평균 수익률 계산
print(f"SPY(S&P500) 연평균수익률(10년): {average_annual_return_spy * 100}%")
print(f"MBB(모기지) 연평균수익률(10년): {average_annual_return_mbb * 100}%")
```

```python
# MBB 배당금/배당수익률
MBB = yf.Ticker('MBB')
dividends = pd.DataFrame()

dividends = MBB.dividends
dividend_mean = round(MBB.dividends.mean(), 2)
dividendYield = MBB.info.get("yield") * 100
dividen_payoutRatio = MBB.info.get('payoutRatio')
dividen_rate = MBB.info.get('dividendRate')

print(f"MBB(모기지) 배당금 : {dividends}\n")
print(f"월평균 배당금(2007~) : ${dividend_mean}")
print(f"배당수익률(전월) : {dividendYield}%")
print(f"순이익 대비 배당금 비율(dividends_payout_ratio): {dividen_payoutRatio}")
```
